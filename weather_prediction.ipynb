```python
x=data.district
plt.bar(x,data.fatalities,width=0.2)#plotting sepal length against species 
plt.xlabel('districts')
plt.ylabel("fatalities")
plt.show()
```

```python
X = data.district


  
X_axis = nm.arange(len(X))
  
plt.bar(X_axis - 0.2, data.actual_rainfall_in_mm, 0.4, label = 'Actual rainfall')
plt.bar(X_axis + 0.2, data.normal_rainfall_in_mm, 0.4, label = 'Normal rainfall')
  
plt.xticks(X_axis, X)
plt.xlabel("districts")
plt.ylabel("rainfall")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
